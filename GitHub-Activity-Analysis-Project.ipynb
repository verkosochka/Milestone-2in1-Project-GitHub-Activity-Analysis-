<h3 style='font-family: Garamond';><b>Data Dictionary and Features</b></h3>
<hr></hr>
<p style='font-family: Garamond ; font-size: 18px;'>Although the full data dictionary is provided in the <code>GitHub_Analysis_BigQuery_Dataset_Implementation.md</code> file, for convenience, it has also been included in this working notebook: 30000 samples of unique repositories with 14 columns.</p>

<ol style='font-family: Garamond ; font-size: 16px;'>
    <li><code>repo_name</code>(STRING, Nullable)</li>
        <ul>
            <li><b>Description:</b> the name of the repository.</li>
            <li><b>Example:</b><code>'tensorflow/tensorflow'</code>, <code>'microsoft/vscode'</code>.</li>
        </ul><br>
    <li><code>language</code>(STRING, Nullable)</li>
        <ul>
            <li><b>Description:</b> the programming language used in the repository (this can represent the language for a specific file or general project use).</li>
            <li><b>Example:</b> <code>'Python'</code>, <code>'JavaScript'</code>.</li>
        </ul><br>
    <li><code>language_bytes</code>(INTEGER, Nullable)</li>
    <ul>
        <li><b>Description:</b> the total number of bytes of code in the repository that is written in the specified <code>language</code>. Represents the size of the code base in terms of bytes.</li>
        <li><b>Example:</b> <code>250000</code>, <code>134500</code>.</li>
    </ul><br>
    <li><code>license</code>(STRING, Nullable)</li>
        <ul>
            <li><b>Description:</b> the license type for the repository, which governs the legal terms for usage and distribution of the code.</li>
            <li><b>Example:</b> <code>'MIT'</code>, <code>'GPL-3.0'</code>.</li>
        </ul><br>
    <li><code>total_files</code>(INTEGER, Nullable)</li>
        <ul>
            <li><b>Description:</b> The total number of files in the repository.</li>
            <li><b>Example:</b> <code>150</code>, <code>320</code>.</li>
        </ul><br>
    <li><code>total_commits</code>(INTEGER, Nullable)</li>
        <ul>
            <li><b>Description:</b> the total number of commits made to the repository.</li>
            <li><b>Example:</b> <code>1200</code>, <code>8000</code>.</li>
        </ul><br>
    <li><code>used_languages</code>(INTEGER, Nullable)</li>
        <ul>
            <li><b>Description:</b> the number of distinct programming languages used in the repository.</li>
            <li><b>Example:</b> <code>2</code>, <code>3</code>.</li>
        </ul><br>
    <li><code>top_language</code>(STRING, Nullable)</li>  
        <ul>
            <li><b>Description:</b> the primary or most used programming language in the repository.</li>
            <li><b>Example:</b> <code>'C#'</code>, <code>'Solidity'</code>.</li>
        </ul><br>
    <li><code>avg_bytes_per_file</code>(FLOAT, Nullable)</li>
        <ul>
            <li><b>Description:</b> the average size of files in the repository, calculated as the total size (in bytes) divided by the number of files.</li>
            <li><b>Formula:</b> <code>avg_bytes_per_file = language_bytes / total_files</code>.</li>
            <li><b>Example:</b> <code>1700.5</code>, <code>420.75</code>.</li>
        </ul><br>
    <li><code>avg_commits_per_file</code>(FLOAT, Nullable)</li>
        <ul>
            <li><b>Description:</b> the average number of commits per file in the repository.</li>
            <li><b>Formula:</b> <code>avg_commits_per_file = total_commits / total_files</code>.</li>
            <li><b>Example:</b> <code>8.5</code>, <code>25.3</code>.</li>
        </ul><br>
    <li><code>commit_intensity_score</code>(FLOAT, Nullable)</li>
        <ul>
            <li><b>Description:</b> a measure of commit intensity, calculated as the total number of commits divided by the total bytes of code in the repository. This metric indicates how active a repository is relative to its size.</li>
            <li><b>Formula:</b> <code>commit_intensity_score = total_commits / language_bytes</code>.</li>
            <li><b>Example:</b> <code>0.005</code>, <code>0.03</code>.</li>
        </ul><br>
    <li><code>avg_files_per_lang</code>(FLOAT, Nullable)</li>
        <ul>
            <li><b>Description:</b> the average number of files per programming language in the repository.</li>
            <li><b>Formula:</b> <code>avg_files_per_lang = total_files / used_languages</code>.</li>
            <li><b>Example:</b> <code>75</code>, <code>50</code>.</li>
        </ul><br>
    <li><code>code_density</code>(FLOAT, Nullable)</li>
        <ul>
            <li><b>Description:</b> the ratio of the bytes of code written in a specific language to the total bytes of code across all languages in the repository.</li>
            <li><b>Formula:</b> <code>code_density = language_bytes / SUM(language_bytes) OVER (PARTITION BY repo_name)</code>.</li>
            <li><b>Example:</b> <code>0.4</code>, <code>0.6</code>.</li>
        </ul><br>
    <li><code>avg_commits_per_lang</code>(FLOAT, Nullable)</li>
        <ul>
            <li><b>Description:</b> the average number of commits for each language used in the repository.</li>
            <li><b>Formula:</b> <code>avg_commits_per_lang = total_commits / used_languages</code>.</li>
            <li><b>Example:</b> <code>300</code>, <code>1000</code>.</li>
        </ul>
</ol>

<h3 style='font-family: Garamond;'><b>Designing New Classification Categories</b></h3>
<hr></hr>

<p style='font-family: Garamond ; font-size: 16px;'>Additionally, we will create new categorical columns: <code>license_type</code> (Open-source vs Non-open source) based on the <code>license</code>, and <code>lang_category</code> (Multilingual, Trilingual, Bilingual, Monolingual) based on the <code>used_languages</code> in the project. It is important to note that all columns with categorical values will subsequently be converted to <code>category</code> type for memory usage optimization.</p>

```python
def category_mapping_used_lang(value: Union[int, float]) -> str:
    """
    Maps the number of languages used in a repository to a corresponding category.

    Parameters
    ----------
    value : Union[int, float]
        The number of languages used in the repository. The value should be a positive integer 
        representing the number of languages used. Valid values are:
        - 1 for 'Monolingual'
        - 2 for 'Duolingual'
        - 3 for 'Trilingual'
        - Any other value will return 'Multilingual'.

    Returns
    -------
    str
        A string representing the category of language usage:
        - 'Monolingual' for a single language,
        - 'Duolingual' for two languages,
        - 'Trilingual' for three languages,
        - 'Multilingual' for more than three languages.
    """
    if value == 1:
        return 'Monolingual'
    elif value == 2:
        return 'Duolingual'
    elif value == 3:
        return 'Trilingual'
    else:
        return 'Multilingual'
```

```python
def category_mapping_license(value: str) -> str:
    """
    Maps a license type to either 'Open-Source' or 'Non-Open Source'.

    Parameters
    ----------
    value : str
        The license type of a repository. Accepted values are:
        - 'unlicense' or 'cc0-1.0' are categorized as 'Non-Open Source'.
        - Any other license will be categorized as 'Open-Source'.

    Returns
    -------
    str
        A string indicating whether the license is 'Open-Source' or 'Non-Open Source'.
        - 'Non-Open Source' for 'unlicense' or 'cc0-1.0'.
        - 'Open-Source' for any other license type.
    """
    if value == 'unlicense' or value == 'cc0-1.0':
        return 'Non-Open Source'
    else:
        return 'Open-Source'
```

```python
def convert_columns_to_category(df: pd.DataFrame, columns: List[str]) -> None:
    """
    Converts specified columns in a DataFrame to the 'category' dtype.

    Parameters
    ----------
    df : pandas.DataFrame
        The DataFrame containing the columns to convert.
    columns : list of str
        A list of column names in the DataFrame that should be converted to 'category' dtype.

    Returns
    -------
    None
        The function modifies the DataFrame in place and does not return a value.

    Raises
    ------
    ValueError
        If any of the columns in the list do not exist in the DataFrame.
    """
    for col in columns:
        if col in df.columns:
            try:
                df[col] = df[col].astype('category')
            except Exception as e:
                print(f"Error converting column {col} to category: {e}")
        else:
            print(f"Column {col} not found in DataFrame.")
```

<h2 style='font-family: Garamond';><b>Code Density Analysis across License Types</b></h2>
<hr></hr>
<p style='font-family: Garamond ; font-size: 18px;'>A Mann-Whitney U Test is applied to compare the mean code density between open-source and licensed (non-open-source) repositories. This statistical test aims to evaluate whether the observed difference in code density is statistically significant with respect to license classification.</p>
<p style='font-family: Garamond ; font-size: 18px;'><b>Here you can find some research papers and videos about the Mann-Whitney U Test (a non-parametric test) due to the non-normality of distributions, ANOVA and the concept of power analysis:</b></p>
<ol style='font-family: Garamond ; font-size: 18px;'>
    <li><a href='https://www.youtube.com/watch?v=Twk6lBhBl88'>The Mann-Whitney U Test: formula and examples.</a></li>
    <li><a href='https://www.scribbr.com/statistics/one-way-anova/'>One-way ANOVA - <b>A</b>nalysis <b>O</b>f <b>Va</b>riance.</a></li>
    <li><a href='https://www.youtube.com/watch?v=0Vj2V2qRU10&list=PLIeGtxpvyG-KA-BLkL391X__r0kU4_hm5'>ANOVA - a visual introduction.</a></li>
    <li><a href='https://www.youtube.com/watch?v=VX_M3tIyiYk'>Power analysis - extremely easy explanation (why it is sugnificant).</a></li>
    <li><a href='https://www.researchgate.net/publication/367461591_Power_analysis_guide'>Power analysis guide.</a></li>
</ol>

```python
data_open_source_df = github_activity_df[github_activity_df['license_type']=='Open-Source']
data_non_open_source_df = github_activity_df[github_activity_df['license_type']=='Non-Open Source']
```

```python
# the Mann-Whitney U test
data_open_source = np.array(data_open_source_df['code_density'])
data_non_open_source = np.array(data_non_open_source_df['code_density'])

print(f"The mean value of open-sourсe repositories' code density is: {np.mean(data_open_source)}.\nThe mean value of non-open source repositories' code density is: {np.mean(data_non_open_source)}.")
```

<p style='font-family: Garamond ; font-size: 18px;'><b>Here we have the following hypothesis for the statistical the Mann-Whitney U test and ANOVA -  One-Way ANOVA (Variance Analysis)):</b></p>

<ol style='font-family: Garamond ; font-size: 18px;'>
    <li><b>$H_0$:</b> The difference in the average code density between open-source and non-open-source repositories is not statistically significant, meaning there is no significant difference in the code density between the two types of repositories.</li>
    <li><b>$H_1$:</b> The difference in the average code density between open-source and non-open-source repositories is statistically significant, indicating that the code density of non-open-source repositories is higher than that of open-source repositories.</li>
</ol>

```python
stat_mannwhitney, p_value_mannwhitney = mannwhitneyu(data_open_source, data_non_open_source, alternative='two-sided')

print(f'The z-statistics value is: {stat_mannwhitney}.\nThe p-value is: {p_value_mannwhitney}.')
```

```python
if p_value_mannwhitney < ALPHA:
    print('The null hypothesis H₀ is rejected: the difference in the average code density between open-source and non-open-source repositories is statistically significant, indicating that the code density of non-open-source repositories is higher than that of open-source repositories.')
else:
    print('The null hypothesis H₀ is not rejected: the difference in the average code density between open-source and non-open-source repositories is statistically significant, indicating that the code density of non-open-source repositories is higher than that of open-source repositories.')
```

<p style='font-family: Garamond ; font-size: 18px;'><b>The implementation of ANOVA:</b></p>

```python
f_stat, p_value_anova = f_oneway(data_open_source, data_non_open_source)

print("F-statistic:", f_stat)
print("P-value:", p_value_anova)

# ANOVA hypothesis testing
if p_value_anova < ALPHA:
    print('The null hypothesis H₀ is rejected: there is a significant difference in code density across the license types.')
else:
    print('The null hypothesis H₀ is not rejected: there is no significant difference in code density across the license types.')
```

<p style='font-family: Garamond ; font-size: 18px;'>To ensure the reliability of the statistical test results, we will perform a <b>power analysis</b> - to determine the probability of correctly rejecting the null hypothesis when it is false, based on the sample size, effect size, and significance level.</p>

```python
mean_open_source = np.mean(data_open_source)
mean_non_open_source = np.mean(data_non_open_source)
std_open_source = np.std(data_open_source)
std_non_open_source = np.std(data_non_open_source)

# Calculate Cohen's d (Effect size)
d = (mean_open_source - mean_non_open_source) / np.sqrt((std_open_source ** 2 + std_non_open_source ** 2) / 2)

# The sample size for both open-source and non-open-source repositories
sample_size = len(data_open_source) + len(data_non_open_source)  

# Perform power analysis
power_analysis = sms.TTestIndPower()
power = power_analysis.solve_power(effect_size=d, nobs1=sample_size, alpha=0.05, ratio=1, alternative='two-sided')

# Print the power result
print(f'Power of the test: {power:.4f}. The sample size is: {sample_size}.')
```

```python
# Calculate the required sample size for 80% power to be confident in the result of power analysis
required_sample_size = power_analysis.solve_power(effect_size=d, power=0.8, alpha=0.05, ratio=1, alternative='two-sided')

print(f"Required sample size for 80% power: {required_sample_size:.0f}.")
```

<p style='font-family: Garamond ; font-size: 18px;'>As we can see, the probability of the correctness of our statistical test results for hypothesis testing is 100%, as confirmed by the power analysis, which shows that for 80% power, 505 samples are required, whereas our dataset contains 30000 samples.</p>

<p style='font-family: Garamond ; font-size: 18px;'><b>Hypotheses for the A/B Test:</b></p>

<ol style='font-family: Garamond ; font-size: 18px;'>
    <li><b>$H_0$</b>: The transition from Python to Java does not affect the <code>commit_intensity_score</code> of repositories.</li>
    <li><b>$H_1$</b>: The transition from Python to Java does affect the <code>commit_intensity_score</code> of repositories. This may be the result of the development of new libraries, frameworks, or the increased popularity of these languages for specific tasks.</li>
</ol>

<p style='font-family: Garamond ; font-size: 18px;'><b><u>NOTE 1 !</u></b> This is an artificial experiment, and we are illustrating the theory of experimental design. Furthermore, through power analysis, it is demonstrated that the sample size in our dataset is NOT sufficient for a robust and reliable hypothesis test and validation (need: <code>36784</code>, have: <code>2686</code> (Python), <code>2172</code> (Java)). However, by calculating the proportion effect size (Cohen's h), we obtain a positive result, indicating a favorable transition from the control group to the treatment group. This suggests that the process of switching from one backend strategy to another, in the context of the <code>commit_intensity_score</code>, is evident — specifically, from Python-based frameworks to Java technologies. We hypothesize that frameworks and project architecture trends are evolving more rapidly in the Java ecosystem.</p>

<p style='font-family: Garamond ; font-size: 18px;'><b><u>NOTE 2 !</u></b> To compare the two groups (Python and Java), <a href='https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mannwhitneyu.html'>the Mann-Whitney U test</a> can be used. <h2 style='font-family: Garamond';><b>ANOVA: Language Diversity vs Code Density</b></h2>
<hr></hr>

<p style='font-family: Garamond ; font-size: 18px;'>Analysis of Variance (ANOVA) is employed to test whether the average code density significantly differs across projects grouped by language usage categories.</p>

<p style='font-family: Garamond ; font-size: 18px;'><b>Let's go back to ANOVA with these Hypotheses</b>:</p>
<ol style='font-family: Garamond ; font-size: 18px;'>
    <li><b>$H_0$</b>: the difference in <code>code_density</code> between language categories is not statistically sagnificant.</li>
    <li><b>$H_1$</b>: there is a statistically significant difference in <code>code_density</code> between language categories.</li>
</ol>

```python
hue_params = filtered_github_activity_df.groupby('lang_category', observed=True)['code_density'].mean().reset_index()
hue_params = hue_params.rename(columns={'code_density': 'mean_code_density'})

filtered_github_activity_df_merged_hue = filtered_github_activity_df.merge(hue_params, on='lang_category')

groups = [filtered_github_activity_df[filtered_github_activity_df['lang_category'] == category]['code_density'] for category in filtered_github_activity_df['lang_category'].unique()]
f_stat_cd, p_value_anova_cd = f_oneway(*groups)

print(f"ANOVA F-statistic: {f_stat_cd}, p-value: {p_value_anova_cd}")

if p_value_anova_cd < 0.05:
    print('The null hypothesis H₀ is rejected: there is a statistically significant difference in code density between language categories.')
else:
    print('The null hypothesis H₀ is not rejected: there is no significant difference in code density between language categories.')

plt.figure(figsize=(10, 6))
sns.boxplot(x='lang_category', y='code_density', data=filtered_github_activity_df_merged_hue, hue='mean_code_density', palette='coolwarm')

plt.title('Code Density by Language Category (Colored by Mean Code Density)')
plt.xlabel('Language Category')
plt.ylabel('Commit Density Score')

plt.xticks(rotation=45)
plt.grid(linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show();
```
